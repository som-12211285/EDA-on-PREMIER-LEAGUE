# 1. Handling missing values
#### Drop columns with too many missing values (like 'Time', 'MarketMax', etc.)

```python
df = pd.read_csv('/content/sample_data/PremierLeague.csv')
```

```python

# 1. Handling missing values
# Drop columns with many missing values
df_cleaned = df.drop(columns=['Time', 'MarketMaxHomeTeam', 'MarketMaxDraw',
                              'MarketMaxAwayTeam', 'MarketAvgHomeTeam',
                              'MarketAvgDraw', 'MarketAvgAwayTeam',
                              'MarketMaxOver2.5Goals', 'MarketMaxUnder2.5Goals',
                              'MarketAvgOver2.5Goals', 'MarketAvgUnder2.5Goals'])
```

```python
# Fill missing numerical values with  the median
numerical_cols = df_cleaned.select_dtypes(include=np.number).columns
df_cleaned[numerical_cols] = df_cleaned[numerical_cols].fillna(df_cleaned[numerical_cols].median())
```

```python

# Convert 'Date' column to datetime
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], format='%d/%m/%y')
# Changed format to '%d/%m/%y' to handle two-digit years
```

```python
# 3. Remove duplicates
df_cleaned.drop_duplicates(inplace=True)
```

```python
# 4. Verify cleaned data
print(df_cleaned.info())
print(df_cleaned.head())
```

# **Step 2: Feature Engineering**

### Adding goal difference and total goals per match

```python
df_cleaned['GoalDifference'] = df_cleaned['FullTimeHomeTeamGoals'] - df_cleaned['FullTimeAwayTeamGoals']
df_cleaned['TotalGoals'] = df_cleaned['FullTimeHomeTeamGoals'] + df_cleaned['FullTimeAwayTeamGoals']
```

```python
# Convert categorical columns (HomeTeam, AwayTeam, FullTimeResult) to numeric using Label Encoding
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df_cleaned['HomeTeamEncoded'] = labelencoder.fit_transform(df_cleaned['HomeTeam'])
df_cleaned['AwayTeamEncoded'] = labelencoder.fit_transform(df_cleaned['AwayTeam'])
df_cleaned['FullTimeResultEncoded'] = labelencoder.fit_transform(df_cleaned['FullTimeResult'])
```

# **Step 3: Multicollinearity Check**

```python
plt.figure(figsize=(10, 8))
corr_matrix = df_cleaned.select_dtypes(include=np.number).corr()  # Select only numeric columns for correlation
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# **Step 4: Feature Scaling**

```python
# Scaling the data for PCA and t-SNE analysis
!pip install scikit-learn #installing scikit-learn which includes StandardScaler
from sklearn.preprocessing import StandardScaler # Importing StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df_cleaned[['HomeTeamEncoded', 'AwayTeamEncoded',
                                            'FullTimeHomeTeamGoals', 'FullTimeAwayTeamGoals',
                                            'GoalDifference', 'TotalGoals']])
```

# **Step 5: Principal Component Analysis (PCA)**

```python
# Installing and Importing PCA
!pip install scikit-learn
from sklearn.decomposition import PCA #Importing PCA from sklearn.decomposition

# Now you can use PCA as before
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_df)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(pca_components[:, 0], pca_components[:, 1], c=df_cleaned['FullTimeResultEncoded'], cmap='viridis', alpha=0.5)
plt.title('PCA of Premier League Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Match Result')
plt.show()
```

# **Step 6: t-SNE for Dimensionality Reduction**

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
tsne_components = tsne.fit_transform(scaled_df)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(tsne_components[:, 0], tsne_components[:, 1], c=df_cleaned['FullTimeResultEncoded'], cmap='coolwarm', alpha=0.5)
plt.title('t-SNE of Premier League Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Match Result')
plt.show()
```

## 🔍 Key Insights from Visualizations

After performing thorough EDA and visualizing various aspects of the Premier League dataset, here are the major insights obtained:

### 📈 1. Correlation Heatmap
- Strong positive correlation between **shots on target** and **goals scored**, indicating shooting accuracy plays a key role.
- A mild negative correlation between **fouls committed** and **points earned**, suggesting that disciplined teams often perform better.

### 📦 2. Boxplots
- **Goals scored** have a few significant outliers — indicating occasional high-scoring matches or standout players.
- **Yellow cards** are more consistently distributed, with most teams falling within a typical range, but a few overly aggressive teams stand out.

### 🏟️ 3. Home vs Away Performance
- Home teams tend to win more matches than away teams, confirming the traditional **home advantage** trend.
- Away teams have slightly fewer goals on average.

### 🧍 4. Player-Level Analysis
- Top goal scorers also lead in shots taken and minutes played, highlighting the importance of regular playtime.
- A few players accumulate high assists with fewer goals, indicating their role as creative playmakers.

### 🛑 5. Defensive Statistics
- Teams with higher **clearances and tackles** often finish mid-table — showing that strong defense alone may not ensure top position.
- **Goalkeepers' saves** were highest in lower-ranked teams, indicating defensive pressure and fewer clean sheets.
